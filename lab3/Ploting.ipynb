```python
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
from scipy.interpolate import interp1d
from scipy.interpolate import CubicSpline
```

```python
x = [3,1,7,3,1,8,10,3,9,11,3,7,8,1,4,2]
y = [2,5,2,3,6,8,12,7,5,10,6,7,9,3,5,5]

x = np.array(x)
y = np.array(y)

m=0.5
b=3.3

plt.plot(x,y,'.')
ybaru=m*x+b
plt.plot (x, ybaru, 'r-')

#plt.plot(x,y,".")
#plt.plot([1,2],"-")
#plt.plot(x,y,"o")
plt.xlabel("Titik X")
plt.ylabel("Titik Y")
plt.title("Diagram X Y")
plt.show()
```

```python
A = np.vstack([x,np.ones(len(x))]).T
# A itu semua titik x di turunin kebawah dan disampingnya ada angka satu
# fungsinya buat perkalian matriks biar ga nol
# .T itu artinya di transpose
#print(A)

print("\n")
Y= np.vstack(y)
#print(Y)
# y ga perlu tambahin angka satunya, y dibalikin biar bisa dikaliin

hasilInvers = np.linalg.inv(np.dot(A.T,A)) 

beta = np.dot(np.dot(hasilInvers, A.T),Y)

print(beta)

m = beta[0]
b = beta[1]

plt.plot(x,y,'.')
ybaru=m*x+b
plt.plot (x, ybaru, 'r-')

plt.xlabel("Titik X")
plt.ylabel("Titik Y")
plt.title("Diagram XÂ Y")
plt.show()
```

```python
# linear interpolation
x=[1,6,11,16]
y=[2,5,3,7]
x=np.array(x)
y=np.array(y)

x_cari=4
f1=interp1d(x,y,kind='linear')
y_cari=f1(x_cari)

plt.plot(x,y,'o-')
plt.plot(x_cari,y_cari,'ro')

# plt.Annotation('Titik 1', xy=(1,2), xytext=(1,2))
# plt.Annotation('Titik 2', xy=(6,5), xytext=(6,5))
# plt.Annotation('Titik 3', xy=(11,3), xytext=(11,3))
# plt.Annotation('Titik 4', xy=(16,7), xytext=(16,7))

plt.xlabel("Titik X")
plt.ylabel("Titik Y")
plt.title("Diagram X Y")
plt.show()
```

```python
# cubic spline
f2=CubicSpline(x,y,bc_type='natural')
x_new=np.linspace(x[0],x[-1:],100000)
y_new=f2(x_new)

x_cari=4
y_cari=f2(x_cari)

plt.plot(x_new,y_new,'-')
plt.plot(x_cari,y_cari,'ro')

plt.show()
```
