```python
import numpy as np
import matplotlib.pyplot as plt
from sympy import *
```

```python
a=symbols('a')
f1=7*a**4-3*a**2+a
df1=diff(f1,a)
df1=lambdify(a,df1,'numpy')
f1=lambdify(a,f1,'numpy')

f2=8*a**6+5*a**4-2
df2=diff(f2,a)
df2=lambdify(a,df2,'numpy')
f2=lambdify(a,f2,'numpy')

print(df1(1))



```

```python
#newthon raphson
def newthon_raphson(f,df,tolerance,max,x0):
    if max==0:
        print('max iteration reached')
        return false
    if np.abs(f(x0))<tolerance:
        print(x0)
        return true
    else:
        return newthon_raphson(f,df,tolerance,max-1,x0-f(x0)/df(x0))
    
```

```python
print(newthon_raphson(f1,df1,0.0001,5,1))
print(newthon_raphson(f2,df2,0.006541,5,1))
```
